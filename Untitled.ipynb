The Datasset is highly unbalanced

0 -- Normal Transaction

1 -- Fradulent Transaction

```python
#Seperating the data for analysis
legit = card_df[card_df.Class == 0]
fraud = card_df[card_df.Class == 1]

#Viewing the shape of the two new df
print(legit.shape, fraud.shape)
```

```python
#Statistical Summary of the amount column of the legitimate df
legit.Amount.describe()
```

```python
#Statistical Summary of the amount column of the fraudlent df
fraud.Amount.describe()
```

## Data Preprocessing

I will be using a Method known as Undersampling to build a sample dataset with similar distribution to the original dataset 

### Number of Frad Transactions -- 492

```python
#Legit Sample
l_samples = legit.sample(n=492)
```

```python
#Concatenating two dataframes
card_new = pd.concat([l_samples, fraud], axis = 0)
```

```python
#Viewing the first five rows of the new df
card_new.head()
```

```python
#Viewing the tail of the df
card_new.tail()
```

```python
#Viewing the distribution of the target value in the new dataframe
card_new.Class.value_counts()
```

```python
#Compare the mean values for both classes of transaction
card_new.groupby('Class').mean()
```

```python
#Splitting the data into features and labels
X = card_new.drop('Class', axis = 1).values
y = card_new.Class.values

#Viewing the labels and features
print(X,y)
```

```python
#Splitting the data into training and test data
SEED = 234
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size = 0.25, stratify = y, random_state = SEED)
```

```python
#Validating the split
print(X.shape, X_train.shape, X_test.shape)
print(y.shape, y_train.shape, y_test.shape)
```

## Model Building

```python
#Creating a logistic regression instance
lr = LogisticRegression()
```

```python
#Training the model 
lr.fit(X_train,y_train)
```

## Model Evaluation

```python
#Training prediction
y_train_pred = lr.predict(X_train)

#Finding the accuracy of the model on the training data
train_acc = accuracy_score(y_train_pred, y_train)

#training accuracy
print(f"The Model Training accuracy is: {train_acc}")
```

```python
#Test Prediction
y_test_pred = lr.predict(X_test)

#Finding the accuracy of the model on the test data
test_acc = accuracy_score(y_test_pred, y_test)

#test accuracy
print(f"The Model Test accuracy is: {test_acc}")
```
